```python
import pandas as pd
import json
from langchain_core.documents import Document

# 1. 꼬인 CSV를 먼저 정제해서 읽기 (on_bad_lines 사용)
df = pd.read_csv('real-estate.csv', encoding='utf-8-sig', on_bad_lines='skip')
df = df.fillna("")
df.columns = [col.replace('\n', ' ').strip() for col in df.columns]

documents = []

for _, row in df.iterrows():
    # 전체 데이터를 dict로 변환 (JSON 형태)
    full_data_dict = row.to_dict()
    
    # [핵심] 검색 모델(Embedding)이 잘 찾을 수 있게 핵심 정보만 문장으로 요약
    # 56개 칼럼을 다 읽게 하면 오히려 노이즈가 생길 수 있습니다.
    summary_text = (
        f"빌딩명: {full_data_dict.get('빌딩명', '')}, "
        f"지역: {full_data_dict.get('지역', '')}, "
        f"현황: {full_data_dict.get('현황', '')}, "
        f"용도: {full_data_dict.get('용도', '')}"
    )
    
    # 실제 DB에 넣을 때는 '요약 텍스트' + '전체 JSON 데이터'를 합칩니다.
    page_content = f"요약: {summary_text}\n\n상세데이터(JSON): {json.dumps(full_data_dict, ensure_ascii=False)}"
    
    doc = Document(
        page_content=page_content,
        metadata={
            "building_id": full_data_dict.get('빌딩명', 'unknown'),
            "price": full_data_dict.get('매도호가(현재)', 0)
        }
    )
    documents.append(doc)

print(documents)
```

```python
import getpass
import os

from pinecone import Pinecone
from langchain_pinecone import PineconeVectorStore
from dotenv import load_dotenv

load_dotenv()

pinecone_api_key = os.environ.get("PINECONE_API_KEY")
print(f'pinecone_api_key == {pinecone_api_key}')


embedding = OpenAIEmbeddings(model='text-embedding-3-large')

pc =Pinecone()

index_name = "bookcity-real-estate"

index = pc.Index(index_name)
vectorstore = PineconeVectorStore(embedding=embedding, index=index)
```

```python

database = PineconeVectorStore.from_documents(documents, model, index_name=index_name)
```

```python
query='파주출판단지 부동산 매물들에 대해 알려줘'
retrieved_docs = database.similarity_search(query)
```
