```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model='gpt-4o-mini')
```

```python
from langchain_core.tools import tool

@tool
def add(a: int, b: int) -> int:
    """Add two numbers together."""
    return a + b
@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers together."""
    return a * b

```

```python

llm_with_tools = llm.bind_tools([add, multiply])
```

```python
query = "What is 2 + 3? What is 4 * 5?"

result = llm_with_tools.invoke(query)
result
```

```python
result.tool_calls   
```

```python
from langchain_core.messages import AnyMessage, HumanMessage
from typing import Sequence


human_message = HumanMessage(query)
message_list: Sequence[AnyMessage] = [human_message]
```

```python
ai_message = llm_with_tools.invoke(message_list)
ai_message
```

```python
ai_message.tool_calls
```

```python
message_list.append(ai_message)
```

```python
tool_message1 = multiply.invoke(ai_message.tool_calls[0])
tool_message2 = add.invoke(ai_message.tool_calls[1])
```

```python
message_list.append(tool_message1)
message_list.append(tool_message2)
```

```python
message_list
```

```python
result = llm_with_tools.invoke(message_list)
```
